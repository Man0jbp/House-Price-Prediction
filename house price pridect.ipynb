```python
from sklearn import linear_model
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(df[["area"]], df["price"])
```

```python
import pandas as pd
predicted_price = reg.predict(pd.DataFrame([[1000]], columns=["area"]))
```

```python
print(predicted_price)
```
